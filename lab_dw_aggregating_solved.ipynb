```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
marketing_data = pd.read_csv(url)
```

```python
marketing_data.head()
```

```python
# Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.
claim_amount = marketing_data[(marketing_data['Total Claim Amount'] > 1000) & (marketing_data['Response'] == 'Yes')]
claim_amount.head()
```

```python
# Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. Write your conclusions.
# Filter for customers who responded "Yes"
filtered_data = marketing_data[marketing_data['Response'] == 'Yes']

# Group by policy type and gender, then calculate the average total claim amount
marketing_data_analysis = filtered_data.groupby(['Policy Type', 'Gender'])['Total Claim Amount'].mean()

# Display the results
print(marketing_data_analysis)
```

```python
# Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.
total_customers_by_state = marketing_data.groupby('State')['Customer'].count()
filtered_states = total_customers_by_state[total_customers_by_state > 500]
print(filtered_states)
```

```python
# Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.
customer_lifetime_value = marketing_data.groupby(['Education', 'Gender'])['Customer Lifetime Value'].agg(['max', 'min', 'median'])
print(customer_lifetime_value)
```

```python
marketing_data['Effective To Date'].unique()
```

```python
# The marketing team wants to analyze the number of policies sold by state and month. Present the data in a table where the months are arranged as columns and the states are arranged as rows.
marketing_data['Effective To Date'] = pd.to_datetime(marketing_data['Effective To Date'])
months = marketing_data['Effective To Date'].dt.month
marketing_data['Month'] = months
pivot_table = marketing_data.pivot_table(index='State', columns='Month', values='Number of Policies', aggfunc='sum')
print(pivot_table)
```

```python
# Step 1: Group by State and Month, and sum the Number of Policies
grouped_data = marketing_data.groupby(['State', 'Month'])['Number of Policies'].sum().reset_index()

# Step 2: Calculate the total number of policies for each state
state_totals = grouped_data.groupby('State')['Number of Policies'].sum().reset_index()

# Step 3: Identify the top 3 states with the highest total number of policies
top_states = state_totals.nlargest(3, 'Number of Policies')['State']

# Step 4: Filter the grouped data to include only the top 3 states
top_states_data = grouped_data[grouped_data['State'].isin(top_states)]

# Step 5: Sort the data by total policies (descending) and then by Month
top_states_data = top_states_data.merge(state_totals, on='State', suffixes=('', '_total'))
top_states_data = top_states_data.sort_values(by=['Number of Policies_total', 'Month'], ascending=[False, True])

# Step 6: Drop the total column (optional, for cleaner output)
top_states_data = top_states_data.drop(columns=['Number of Policies_total'])

# Step 7: Display the final DataFrame
print(top_states_data)
```
