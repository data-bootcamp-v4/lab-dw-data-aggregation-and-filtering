```python
import pandas as pd

# Load the raw dataset
df = pd.read_csv("marketing_customer_analysis.csv")

# Standardize column names
df.columns = df.columns.str.lower().str.replace(" ", "_")
df.head()
```

```python
# Customers with total claim < $1,000 and response = 'Yes'
low_claim_yes_response = df[(df['total_claim_amount'] < 1000) & (df['response'] == 'Yes')]
low_claim_yes_response.head()
```

```python
# Filter for only "Yes" responders
yes_responders = df[df['response'] == 'Yes']

# Group by policy_type and gender
insights = yes_responders.groupby(['policy_type', 'gender'])[['monthly_premium_auto', 'customer_lifetime_value']].mean().round(2)
insights
```

```python
# Compare with total claim amount in the same groups
claim_patterns = yes_responders.groupby(['policy_type', 'gender'])['total_claim_amount'].mean().round(2)
claim_patterns
```

```python
# Total number of customers per state
customers_by_state = df['state'].value_counts()

# Filter states with more than 500 customers
states_over_500 = customers_by_state[customers_by_state > 500]
states_over_500
```

```python
# Aggregate CLV stats
clv_stats = df.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median']).round(2)
clv_stats
```

```python
# Ensure date column is datetime type
df['effective_to_date'] = pd.to_datetime(df['effective_to_date'])

# Extract month name
df['month'] = df['effective_to_date'].dt.month_name()

# Count policies by state and month
policies_state_month = df.groupby(['state', 'month']).size().unstack().fillna(0).astype(int)
policies_state_month
```

```python
# Total policies sold per state
total_sales_by_state = df['state'].value_counts()

# Get top 3 states
top_3_states = total_sales_by_state.head(3).index.tolist()

# Filter and count
top_3_data = df[df['state'].isin(top_3_states)]
top_3_summary = top_3_data.groupby(['state', 'month']).size().unstack().fillna(0).astype(int)
top_3_summary
```

```python
# Group by marketing channel and response
response_rate = df.groupby(['sales_channel', 'response']).size().unstack().fillna(0)

# Calculate response percentage
response_rate['response_rate_%'] = (response_rate['Yes'] / response_rate.sum(axis=1) * 100).round(2)
response_rate
```
