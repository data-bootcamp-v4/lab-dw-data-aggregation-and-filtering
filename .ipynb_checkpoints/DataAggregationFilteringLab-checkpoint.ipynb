```python
import pandas as pd
import numpy as np
```

```python
marketing_customer_analysis_cleaned = pd.read_csv("marketing_customer_analysis_cleaned.csv")

marketing_customer_analysis_cleaned
```

```python
print(marketing_customer_analysis_cleaned['State'].unique())
```

```python
# 1
```

```python
high_total_claim_amount_df = marketing_customer_analysis_cleaned[(marketing_customer_analysis_cleaned['Total Claim Amount'] > 1000) & (marketing_customer_analysis_cleaned['Response'] == 'Yes')].copy()
```

```python
high_total_claim_amount_df
```

```python
# 2
```

```python
high_total_claim_amount_df = marketing_customer_analysis_cleaned[marketing_customer_analysis_cleaned['Response'] =='Yes'].copy()

claim_amount_avg = high_total_claim_amount_df.groupby(['Policy Type', 'Gender'])['Total Claim Amount'].mean()

print(claim_amount_avg)
```

```python
# 3
```

```python
customers_by_state = marketing_customer_analysis_cleaned['State'].value_counts()

filtered_customers_by_state = customers_by_state[customers_by_state > 500]

print(filtered_customers_by_state)
```

```python
# 4
```

```python
clv_median = marketing_customer_analysis_cleaned.groupby(['Education', 'Gender'])['Customer Lifetime Value'].median()

print(clv_median)
```

```python
# 5
```

```python
max_open_complaints = marketing_customer_analysis_cleaned.groupby(['State', 'Policy Type'])['Number of Open Complaints'].max()

min_open_complaints = marketing_customer_analysis_cleaned.groupby(['State', 'Policy Type'])['Number of Open Complaints'].min()

avg_open_complaints = marketing_customer_analysis_cleaned.groupby(['State', 'Policy Type'])['Number of Open Complaints'].mean()

print("Maximum number of open complaints:")
print(max_open_complaints)
print("\nMinimum number of open complaints:")
print(min_open_complaints)
print("\nAverage number of open complaints:")
print(avg_open_complaints)
```

```python
# 6
```

```python
marketing_customer_analysis_cleaned["Effective To Date"] = pd.to_datetime(marketing_customer_analysis_cleaned["Effective To Date"])

marketing_customer_analysis_cleaned["Month"] = marketing_customer_analysis_cleaned["Effective To Date"].dt.strftime("%b")

policies_sold_pivot = pd.pivot_table(marketing_customer_analysis_cleaned, values='Number of Policies', index='State', columns='Month', aggfunc='sum')

print(policies_sold_pivot)
```

```python
# 7
```

```python
policies_sold_by_state_month = marketing_customer_analysis_cleaned.groupby(['State', 'Month']).size().reset_index(name='Count')

sorted_policies_sold = policies_sold_by_state_month.sort_values(by='Count', ascending=False)

top_3_states = sorted_policies_sold['State'].head(3)

new_dataframe = sorted_policies_sold[sorted_policies_sold['State'].isin(top_3_states)]

print(new_dataframe)
```

```python
# 8
```

```python
unpivoted_data = pd.melt(marketing_customer_analysis_cleaned, id_vars='Sales Channel', value_vars='Response', value_name='Response').dropna()
filtered_data = unpivoted_data[unpivoted_data['Response'] == 'Yes']

response_rate = filtered_data.groupby('Sales Channel').size() / unpivoted_data.groupby('Sales Channel').size()

print(response_rate)
```
