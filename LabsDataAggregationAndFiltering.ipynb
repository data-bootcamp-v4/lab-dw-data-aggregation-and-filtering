```python
import pandas as pd
import numpy as np
```

```python
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv"
df = pd.read_csv(url)
df.head(3)
```

```python
df.shape
```

```python
filtered_df = df[(df['Total Claim Amount'] < 1000) & (df['Response'] == 'Yes')]
filtered_df.head(3)
```

```python
filtered_df.shape
```

```python
responded_df = df[df['Response'] == 'Yes']

summary = responded_df.groupby(['Policy Type', 'Gender'])[
    ['Monthly Premium Auto', 'Customer Lifetime Value', 'Total Claim Amount']
].mean().reset_index()

summary
```

```python
summary.shape
```

```python
### Corporate Auto males are more profitable with high value and little claims and 
### personal auto males have low value and high claims: this makes them riskier.
```

```python
customers_per_state = df['State'].value_counts()

states_over_500 = customers_per_state[customers_per_state > 500]

states_over_500_df = states_over_500.reset_index()
states_over_500_df.columns = ['State', 'Customer Count']
states_over_500_df
```
