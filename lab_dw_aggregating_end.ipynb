1. Create a new DataFrame that only includes customers who:
   - have a **low total_claim_amount** (e.g., below $1,000),
   - have a response "Yes" to the last marketing campaign.

```python
import pandas as pd

url = ('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')
df_1 = pd.read_csv(url)
```

```python
type(df_1)
```

```python
def clean_column_names(df):
    """
    Cleans and standardizes column names in a DataFrame.
    """
    df.columns = [col.lower().replace(' ', '_') for col in df.columns]
    return df
```

```python
# Standardize column names

df_1.columns = df_1.columns.str.replace(' ','_').str.lower()
```

```python
df_1 = df_1.drop(columns=['unnamed:_0'], axis = 1)
```

```python
# Filter dataframe df_1 to DataFrame with 'total_claim_amout' and response 'Yes'
# Subset of data based in specified conditions.

low_claim_df = df_1[(df_1['total_claim_amount'] < 1000) & (df_1['response'] == 'Yes')]
low_claim_df
```

2. Using the original Dataframe, analyze:
   - the average `monthly_premium` and/or customer lifetime value by `policy_type` and `gender` for customers who responded "Yes", and
   - compare these insights to `total_claim_amount` patterns, and discuss which segments appear most profitable or low-risk for the company.

```python
# Filter the dataframe for customers who answer yes

respond_yes = df_1[df_1['response'] == 'Yes'].copy()
```

```python
respond_yes
```

```python
# pivot table to calculate avg monthly_premium and customer_lifetime_value
# for each combination policy_type - gender

pivot_premium = pd.pivot_table(
    respond_yes,
    values=['monthly_premium_auto', 'customer_lifetime_value'],
    index=['policy_type', 'gender'],
    aggfunc='mean'
).round(2)
```

```python
pivot_premium
```

ANALYSIS

Average Monthly Premium and CLV:

Corporate Auto policies, especially for males, show a significantly higher average monthly_premium and customer_lifetime_value, than other policy types. This segment appears to be highly profitable.

Personal Auto policies, for both genders, have lower average monthly premiums but still represent a substantial portion of the business.

Special Auto policies generally have a high customer_lifetime_value, particularly for males, indicating that these customers are valuable over time.

```python
# This is a similar pivot table but for the avg total_claim_amout.
# Helps understand risk associate with each segment - higher claim amount more risk.

pivot_claim = pd.pivot_table(
    respond_yes,
    values='total_claim_amount',
    index=['policy_type', 'gender'],
    aggfunc='mean'
).round(2)
```

```python
pivot_claim
```

ANALYSIS

Average Total Claim Amount:

Corporate Auto policies also have a notably high average total_claim_amount, which suggests this profitable segment might also be high-risk in terms of claims.

Special Auto policies, again with a high CLV, also show a high average total_claim_amount, particularly for males. This confirms that while they are profitable, they are not necessarily low-risk.

Personal Auto policies have the lowest average total_claim_amount across both genders.


Conclusion:

Based on this analysis, the most profitable segment is Corporate Auto and Special Auto, especially for male customers, due to their high customer_lifetime_value and monthly_premium. However, these are also high-risk segments due to their higher average total_claim_amount.

The lowest-risk and still valuable segment is Personal Auto, as they have a significantly lower total_claim_amount. While their profitability might be lower on a per-customer basis, their low-risk profile could make them a stable and reliable customer base.

3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.

```python
# Count the number of customers in each state:

customers_state = df_1['state'].value_counts().reset_index()

# Change columns names to clarity:

customers_state.columns = ['state', 'total_customers']

customers_state
```

```python
# create a filter to apply and retrieve only states with more than 500 customers:

filter_customers_state = customers_state[customers_state['total_customers'] > 500]

filter_customers_state
```

ANALYSIS

The final table clearly shows which states are the company's largest markets, with California, Oregon, and Arizona having significantly more than 500 customers. This is crucial information for business strategy, marketing resource allocation, and operational planning. The marketing team can use this to focus campaigns, the sales team can use it to determine staffing needs, and the customer service department can use it to anticipate call volumes in these high-volume states.

4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.

```python
# Clean and standardize the categorical columns 'education' and 'gender'.
# Standardize 'education' to Title case and 'gender' to single letters.
df_1['education'] = df_1['education'].str.title()
df_1['gender'] = df_1['gender'].replace({'Male': 'M', 'Female': 'F'})

df_1
```

```python
# Group by education and gender, then calculate max, min, and median CLV:

clv_summary = df_1.groupby(['education', 'gender'])['customer_lifetime_value'].agg(['max', 'min', 'median'])

# Round the values for better readability:

clv_summary = clv_summary.round(2)

clv_summary
```

CONCLUSIONS:

Based on the analysis of the summary table:
High-Value Customers: 
- Customers with a Bachelor's degree have the highest median CLV, especially females. This suggests that while there may be outliers with a higher CLV in other groups, the typical Bachelor's degree holder is more valuable to the company over their lifetime.

CLV Range: 
- The range between the minimum and maximum CLV is substantial across all segments. This indicates a wide variation in customer value, regardless of their education or gender.

Gender Differences: 
- Within each education level, there are some differences in CLV between genders, but the median values are generally closer than the max and min values. This suggests that extreme outliers might be driving the max/min differences, but the typical customer's value is more similar across genders within the same education level.
