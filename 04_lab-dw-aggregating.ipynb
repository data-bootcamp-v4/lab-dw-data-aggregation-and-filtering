```python
# your code goes here
import pandas as pd
import re
```

```python
url = 'https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv'
df_marketing = pd.read_csv(url)
df_marketing
```

```python
df_marketing.shape
```

```python
df_marketing.dtypes
```

```python
#1. filtered data - combined filter (&) ensures both conditions are satisfied
filtered_data = df_marketing[(df_marketing['Total Claim Amount'] > 1000) & (df_marketing['Response'] == 'Yes')]
filtered_data.head()
```

```python
#2. analyze the average total_claim_amount by each policy type

positive_response = df_marketing[df_marketing['Response'] == 'Yes']
average_claims = positive_response.groupby(['Policy Type', 'Gender'])['Total Claim Amount'].mean().reset_index()
average_claims
```

```python
"""
Across all policy types, the average Total Claim Amount is relatively consistent, 
ranging from approximately $408 to $458. Personal Auto policies have the highest 
average claim amounts for both genders, almost the same (Female: $452.97/Male: $457.01)
General Insights:
The gender differences in claim amounts are relatively small.
Personal Auto policyholders, regardless of gender, tend to file higher claims 
on average compared to other policy types. Corporate Auto has the lowest average claims, 
especially for male customers.
Gender Analysis:
In Personal Auto, males have slightly higher claims than females, but the difference is minimal.
In Corporate Auto, females have slightly higher average claims ($433.74) than males ($408.58).
Interestingly, in Special Auto, females also have higher claims ($453.28) compared to males ($429.53).

Focus marketing campaigns on Personal Auto policyholders as they tend to have higher claims, 
indicating more active engagement. Further analysis why Corporate Auto policies, especially for males, 
have lower claims and assess if it aligns with company expectations or needs adjustments.

"""
```

```python
#3. 
customer_counts = df_marketing.groupby('State')['Customer'].count().reset_index()
customer_counts.rename(columns={'Customer': 'Number of Customers'}, inplace=True)
```

```python
filtered_states = customer_counts[customer_counts['Number of Customers'] > 500]
filtered_states
```

```python
#4. 
lifetime_value_stats = df_marketing.groupby(['Education', 'Gender'])['Customer Lifetime Value'].agg(['max', 'min', 'median']).reset_index()

lifetime_value_stats
```
